```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'  # 모든 텍스트 요소에 대해 굵은 글씨 적용
plt.rcParams['axes.labelweight'] = 'bold'  # 축 라벨에 대해 굵은 글씨 적용

# Excel 파일 경로 (실제 환경에서는 적절한 파일 경로로 변경해야 함)
file_path = '/Users/kj/Library/CloudStorage/OneDrive-한림대학교/연구 및 과제/박사논문/Thesis.xlsx'

# Excel 파일 읽기
df = pd.read_excel(file_path)

# 'age' 열에 대한 히스토그램 그리기, bins의 range를 수정합니다.
# 예를 들어, 나이의 최소값과 최대값이 각각 20세와 80세라고 가정할 때:
plt.hist(df['ICU days_before rectal swab (+)'], bins=range(0,490), color='#BDA0CB')
plt.xlabel('ICU days before rectal swab (+)', fontsize=16)
plt.ylabel('Number of Samples', fontsize=16)
plt.xticks(fontsize=14)
plt.savefig('/Users/kj/Library/CloudStorage/OneDrive-한림대학교/연구 및 과제/박사논문/ICU day.tiff', format='tiff', dpi=300)
plt.show()
```

```python
# 성별 데이터를 기반으로 남성과 여성의 수를 계산합니다.
NoBSI_count = sum(df['30_days_label'] == 0)
BSI_count = sum(df['30_days_label'] == 1)

# 성별에 따른 개수를 막대그래프로 표현합니다.
labels = ['NoBSI', 'BSI']
counts = [NoBSI_count, BSI_count]

plt.bar(labels, counts, color='#EA6176', edgecolor='#727171')
plt.xlabel('CRE BSI', fontsize=16)
plt.ylabel('Number of Samples', fontsize=16)
plt.xticks(fontsize=14)
# tiff 형식으로 그래프 저장
plt.savefig('/Users/kj/Library/CloudStorage/OneDrive-한림대학교/연구 및 과제/박사논문/thesis/BSI_distribution.tiff', format='tiff', dpi=300)
plt.show()
```
